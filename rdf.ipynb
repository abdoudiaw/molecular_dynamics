```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

Load the LAMMPS dump file containing the radial distribution functions

```python
data=np.loadtxt('gofr_BIM_CH.csv', skiprows=4, unpack=True)

mpl.rc('font', size = 18)
mpl.rcParams['figure.figsize'] = 12,6
fig, ax = plt.subplots(1,1)

bohr_radius=5.29e-9
ax.plot(data[1]/bohr_radius, data[2], linewidth = 2, label = '$g_{11}$')
ax.plot(data[1]/bohr_radius, data[3], linewidth = 2, label = '$g_{12}$')
ax.plot(data[1]/bohr_radius, data[4], linewidth = 2, label = '$g_{22}$')

ax.set_ylabel(r'radial distribution functions',fontsize=15,labelpad=15)
ax.set_xlabel('r [$a_B$]',fontsize=15)

ax.legend(loc='best');
ax.grid(linestyle='dashed', linewidth='0.5')
plt.tight_layout()
```
